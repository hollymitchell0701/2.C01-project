```python
# merge all the session.laps with the seassion.weather_data with StartLapTime to the closest Time in weather data

# copy session laps
laps = session.laps.copy()

for i, lap in laps.iterrows():
    # find the closest weather data point to the lap start time
    closest_weather = session.weather_data.iloc[(session.weather_data['Time'] - lap['LapStartTime']).abs().argmin()]
    # assign the weather data to the lap
    laps.at[i, 'AirTemp'] = closest_weather['AirTemp']
laps
```
